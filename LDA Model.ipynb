```python
import pandas as pd
import numpy as np
import gensim
from gensim.models import ldamodel
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import pyLDAvis
from pyLDAvis import gensim_models
# from gensim.corpora.dictionary import Dictionary
```

## Read in CSV

```python
twt = pd.read_csv('C:/Users/User/Desktop/Python/Projects/Labour_Party_Conf_Tweets/twt.csv')
```

## Final pre-processing

```python
# -- tokenize

twt['tweet_proc'].dropna
twt['tweet_proc'] = twt['tweet_proc'].astype(str)
```

```python
twt['word_token'] = twt['tweet_proc'].apply(nltk.word_tokenize)

twt['sentence_token'] = twt['tweet_proc'].apply(nltk.sent_tokenize)

tokens = twt['word_token']

tokens
```

```python
# Stemming (reducing words to their stems)

pstem = nltk.stem.PorterStemmer()

twt['stems'] = twt['word_token'].apply(lambda x : [pstem.stem(y) for y in x])

stems = twt['stems']

stems
```

```python
# lemmatizer (reducing words to their stems)

lemm = nltk.stem.WordNetLemmatizer()

twt['lemms'] = twt['word_token'].apply(lambda x : [lemm.lemmatize(y) for y in x])

lemms = twt['lemms']

lemms
```

```python
# -- Vectorization

gensim_dict = gensim.corpora.dictionary.Dictionary(lemms)

bow_corpus = [gensim_dict.doc2bow(token, allow_update=True) for token in lemms]
```

```python
# -- Define LDA model

lda = gensim.models.LdaMulticore(bow_corpus, num_topics=32, id2word=gensim_dict)
```

```python
# -- LDA DataVis
pyLDAvis.enable_notebook()

pyLDAvis.gensim_models.prepare(lda, bow_corpus,gensim_dict)
```
